```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
from scipy.stats import linregress

# read in each csv file into a separate variable
df_population = pd.read_csv('./Resources/Population_Gender.csv',skiprows=0)
df_population
df_electricity = pd.read_csv('./Resources/Electricity_Global.csv',skiprows=0)
df_electricity
```

```python
# merge and melt
df_population_melt = pd.melt(df_population,
                        id_vars=['Country Name', 'Country Code','Series Name','Series Code'], 
                        var_name='Year', value_name='Population_Value')
df_electricity_melt = pd.melt(df_electricity,
                             id_vars=['Country Name', 'Country Code','Series Name','Series Code'],
                             var_name='Year',value_name='Electricity Value')

# Delete columns that is not needed
df_population_melt.drop(['Series Name','Series Code'],axis=1,inplace=True)
df_electricity_melt.drop(['Series Name','Series Code'],axis=1,inplace=True)


# Merge based on common columns
df_merge = df_population_melt.merge(df_electricity_melt,
                               how='outer',
                              on=['Country Name', 'Country Code', 'Year'])

# Sort the country then by year

df_combined = df_merge.sort_values(by=['Country Name','Year'])
df_combined.head()
```

```python
#Clean up the year to remove the values inside[]
df_combined["Year"]
```

```python
sample_string = "2016 [YR2016]"
```

```python
def combinedlist(sample_row):
    sep = "["
    sample_row = str(sample_row)
    one = sample_row.split(sep,1)[0]
    return one
```

```python
pandas_series = df_combined["Year"]
type(pandas_series)
```

```python
dictionary = pandas_series.to_dict()
type(dictionary)
```

```python
df_combined["YEAR"] = df_combined.apply(lambda row: combinedlist(row["Year"]),axis=1)
df_combined_new = df_combined.drop(['Year'], axis=1)
```

```python
df_combined_final =df_combined_new[['Country Name', 'Country Code','YEAR','Population_Value', 'Electricity Value']]
df_combined_final.dropna(inplace=True)
df_combined_final.head(25)
```

```python
#Delete Rows with Null values Country and Country Code
df_combined_complete = df_combined_final.dropna(subset=['Country Name', 'Country Code'])
df_combined_complete
```

```python
#Select SEA Countries only 
SEA =['Indonesia','Philippines','Singapore','Cambodia','Malaysia', 'Thailand', 'Vietnam', 'Timor-Leste', 'Brunei Darussalam', 'Myanmar'] 
  
SEA_df = df_combined_complete.loc[df_combined_complete['Country Name'].isin(SEA)]
SEA_df
```

```python
SEA_df['Population_Value'] = SEA_df['Population_Value'].astype(int)
```

```python
#Create a graph of the above dataframe
SEA_df.plot("Country Name","Population_Value")

#set chart title
plt.title("Population Trend")
plt.xlabel("Country")
plt.ylabel("Population Value")
```

```python
#Extract the combined dataframe to csv
compression_opts = dict(method='zip',
                        archive_name='out.csv')  
df_combined_complete.to_csv('out.zip', index=False,
          compression=compression_opts) 
```
