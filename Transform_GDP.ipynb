```python
#Read dataset
import pandas as pd
import decimal 
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('./Resources/Human_GDP.csv', skiprows=0)
df.drop('Series Code', axis=1, inplace=True)
df.head()
```

```python
#Rename years for example 2020 [YR2020] to Year 2020
df_gdp = df.rename(columns={'2015 [YR2015]': '2015', '2016 [YR2016]':'2016', '2017 [YR2017]': '2017', '2018 [YR2018]': '2018', '2019 [YR2019]':'2019', 'Economy Name':'Country Name'})
df_gdp.head()
```

```python
# Transform data set into long form instead of wide
df_gdp_melt = pd.melt(df_gdp, id_vars=['Country Name', 'Country Code', 'Series Name' ], var_name='Year', value_name='GDP')
df_gdp_melt.head()
```

```python
#Delete Rows with Null values Country Code
df_gdp_melt_final = df_gdp_melt.dropna(subset=['Country Code','Year'])
df_gdp_melt_final
```

```python
#Figure out the data types to identify columns to convert
series = df_gdp_melt.dtypes
series
```

```python
# convert year to a date time
df_gdp_melt['Year'] = pd.to_numeric(df_gdp_melt['Year'])
```

```python
df_gdp_melt.fillna('0', inplace=True)
```

```python
#convert null 
df_gdp_melt[['GDP']] = df_gdp_melt[['GDP']].fillna('0')
```
